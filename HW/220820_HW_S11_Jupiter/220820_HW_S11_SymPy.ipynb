### f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

```python
f, x = sym.symbols('f x') 
f = -12*x**4*sym.sin(sym.cos(x)) - 18*x**3+5*x**2 + 10*x - 30 
display(f)
```

#### 1. Определить корни

```python
sym.solveset(-12*x**4*sym.sin(sym.cos(x)) - 18*x**3+5*x**2 + 10*x - 30, x)
```

```python
sym.solve([x>-100, x<100, f], x) # попытка найти корни в интервале -100 < x < 100
```

#### 2. Найти интервалы, на которых функция возрастает

- производная функции:

```python
f, x = sym.symbols('f x') 
f = -12*x**4*sym.sin(sym.cos(x)) - 18*x**3+5*x**2 + 10*x - 30 
derivative_f = f.diff(x)

derivative_f
```

- Значения X, где производная равна нулю - границы интервалов возрастания и убывания цункции.
- В итервалах, где производная функции > 0, значение функции возрастает.
- Решим уравнение:

```python
sym.solveset(derivative_f, x)
```

- построим график производной:

```python
sym.plot(derivative_f, (x, -100, 100))
```
